# Ex 1

Không dùng sklearn, hãy tự implement Linear Regression bằng numpy để làm các bài sau (lưu ý nếu không fit được ra weight đúng thì có thể do số quá lớn -> scale nhỏ lại)

Fit model parabol linear regression cho dữ liệu:  https://github.com/nttuan8/DL_Tutorial/blob/master/L1/data_square.csv



```python
class LinearRegression():
    def __init__(self,x,y,df):
        self.x = df[x]
        self.y = df[y]
        self.w = None
        self.x_new = df[x]
        self.y_fit = None
    
    def tim_attribute_model_voi_da_thuc_bac_n(self,n):
        # 2 attribute cua model
        self.x_new = np.column_stack([self.x**i for i in range(n+1)])
        #calculate weight
        XT_X = np.dot(self.x_new.T, self.x_new)
        XT_t = np.dot(self.x_new.T, self.y.values.reshape(-1,1))
        self.w = np.dot(np.linalg.pinv(XT_X), XT_t)
        self.y_fit = np.dot(self.x_new,self.w)

    def plot_data_and_fit(self,n):
        self.tim_attribute_model_voi_da_thuc_bac_n(n)
        plt.plot(self.x, self.y, 'ro')
        plt.plot(self.x,self.y_fit)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()


    
    
```

```python
df_1 = pd.read_csv('data_square.csv')
df_1
```

```python
# Dien Tich
X = df_1['Diện tích']
# Price
y = df_1['Giá nhà']
# Visualize data 
plt.plot(X, y, 'ro')
plt.xlabel('Diện tích')
plt.ylabel('Giá')
plt.show()
```

```python
e1 = LinearRegression('Diện tích','Giá nhà',df_1)
e1.plot_data_and_fit(2)
```

```python
#c
class RidgeRegression(LinearRegression):
    def __init__(self,x,y,lamda,df):
        super().__init__(x,y,df)
        self.lamda = lamda
    
    def update_attribute(self,n):
        # cac phep tinh
        self.x_new = np.column_stack([self.x**i for i in range(n+1)])
        XT_X = np.dot(self.x_new.T, self.x_new)
        XT_t = np.dot(self.x_new.T, self.y.values.reshape(-1,1))
        lamdaI = self.lamda * np.eye(self.x_new.shape[0])
        
        # cac attribute trong model
        self.w = np.dot(np.linalg.pinv(XT_X + lamdaI), XT_t)
        self.x_new = np.column_stack([self.x**i for i in range(n+1)])
        self.y_fit = np.dot(self.x_new,self.w)
    
    def plot_update_data(self,n):
        self.update_attribute(n)
        plt.plot(self.x, self.y, 'ro')
        plt.plot(self.x,self.y_fit)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()

    
    
```

```python
c = RidgeRegression('x','y',0.002,df_2)
```

```python
c.update_attribute(9)
c.w
```

```python
c.plot_data_and_fit(9)
c.plot_update_data(9)
```
